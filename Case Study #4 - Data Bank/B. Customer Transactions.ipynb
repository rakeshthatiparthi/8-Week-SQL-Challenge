# **B. Customer Transactions**

<span style="font-size: 14px;"><b>1.What is the unique count and total amount for each transaction type?</b></span>

```sql
SELECT
	txn_type,
	count(txn_type) as unique_count,
	sum(txn_amount) as total_amount
FROM customer_transactions
GROUP BY txn_type
```

<span style="font-size: 14px;"><b>2.What is the average total historical deposit counts and amounts for all customers?</b></span>

```sql
SELECT
	customer_id,
	SUM(CASE
		WHEN txn_type='deposit' THEN 1
	END) as deposit_count,
		SUM(CASE
		WHEN txn_type='deposit' THEN txn_amount
	END) as total_deposit_amount
FROM customer_transactions
GROUP BY customer_id
ORDER BY customer_id
```

<span style="font-size: 14px;"><b>3.For each month - how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?</b></span>

```sql
with cte as(
	SELECT
		customer_id,
		TO_CHAR(txn_date, 'month') as month,
		SUM(CASE
			WHEN txn_type = 'deposit' THEN 1 ELSE 0
		END) as deposit_count,
		SUM(CASE
			WHEN txn_type = 'purchase' THEN 1 ELSE 0
		END) as purchase_count,
		SUM(CASE
			WHEN txn_type = 'withdrawal' THEN 1 ELSE 0
		END) as withdrawal_count
	FROM customer_transactions
	GROUP  BY customer_id,TO_CHAR(txn_date, 'month'))
	SELECT
		month,
		SUM(CASE
			WHEN deposit_count>1 and purchase_count>=1 or withdrawal_count>=1 THEN 1
			ELSE 0
		END) as customer_count
	FROM cte
	GROUP BY month
```

<span style="font-size: 14px;"><b>4. What is the closing balance for each customer at the end of the month?</b></span>

```sql
with cte3 as(
with cte2 as(
with cte1 as(
SELECT 
	customer_id,
	EXTRACT(MONTH FROM txn_date) as month,
	SUM(CASE
			WHEN txn_type = 'deposit' THEN txn_amount ELSE 0
		END) as total_deposit,
		SUM(CASE
			WHEN txn_type = 'purchase' THEN txn_amount ELSE 0
		END) as total_purchase,
		SUM(CASE
			WHEN txn_type = 'withdrawal' THEN txn_amount ELSE 0
		END) as total_withdrawal
FROM customer_transactions
GROUP BY customer_id,EXTRACT(MONTH FROM txn_date))
SELECT
	customer_id ,
	month,
	total_deposit,
	total_purchase,
	total_withdrawal,
	total_deposit-total_purchase-total_withdrawal as month_end_amt
FROM cte1)
SELECT
*,
LAG(month_end_amt,1,'0') OVER(partition by customer_id order by month) as previous_month_end_amt
FROM cte2)
SELECT
customer_id,
month,
total_deposit,
total_purchase,
total_withdrawal,
month_end_amt+previous_month_end_amt as month_end_closing_balance
FROM cte3
```
